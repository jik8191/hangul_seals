```python
from collections import Counter
from sys import argv
import json
from hanziconv import HanziConv



OUTPUT_FILENAME_SIMPL = 'chars_simpl.json'
OUTPUT_FILENAME_TRAD = 'chars_trad.json'

d = {}
with open("3500chars.txt") as f:
    for line in f:
       (_, val, key) = line.split()
       d[int(key)] = val


d_sorted = dict(sorted(d.items(), reverse=True))

# print(d_sorted)

top_chars = [i[1] for i in d_sorted.items()]


top_chars = top_chars[:512]
print("Number of characters written:", len(top_chars))

assert len(top_chars) == 512, "Didn't get 512 characters."
for ele in top_chars:
    assert len(ele) == 1, "found a character that wasn't of length 1"

with open(OUTPUT_FILENAME_SIMPL, 'w') as f:
    json.dump(top_chars, f)
    print("Wrote top 512 simplified characters to", OUTPUT_FILENAME_SIMPL)

print(top_chars)
    
top_chars_trad = [HanziConv.toTraditional(char) for char in top_chars]

print(top_chars_trad)

with open(OUTPUT_FILENAME_TRAD, 'w') as f:
    json.dump(top_chars_trad, f)
    print("Wrote top 512 traditional characters to", OUTPUT_FILENAME_TRAD)


# TODO:
# Traditional dataset
# Create sigils from both simplified and traditional
# with labelling of the Mandarin pronunciation
# Further check for duplicates in General Chinese later


# Create a dictionary of sounds in romanization to characters
# start by converting all of the above chars to sounds using terrapinyin
# ideally find one for middle chinese, it should exist

# use terra pinyin for tone numbers
# create a dictionary with keys as characters, values as the strings
sounds = {}
with open("sounds/terra_pinyin.dict.yaml") as f:
    for i in range(33):
        next(f)
    for line in f:
        line_split = line.split()
        if (len(line_split) == 0):
            break
        else:
            key, val = line.split()[:2]
            sounds[key] = val

print(sounds['是'])
print(HanziConv.toSimplified('繁簡轉換器'))

OUTPUT_FILENAME_DICT = 'chars_trad.txt' 
with open('chars_trad.txt', 'w') as f:
    for char in top_chars[0:255]:
        trad_char = HanziConv.toTraditional(char)
        f.write("%s %s\n" %(sounds[trad_char], trad_char))
        # f.write("%s\n" %( trad_char))
    print("Wrote top 512 characters to", OUTPUT_FILENAME_DICT)
    
```
